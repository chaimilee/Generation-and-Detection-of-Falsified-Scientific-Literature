```python
import csv
filename = 'drive/MyDrive/DSCI 550 Team Project/Assignment 2/image_links2.csv'
writefilename = 'image_links3.csv'
start_link = 'http://0.0.0.0:8764/inception/v3/caption/image?url='
image_link_lst = list()
with open(filename, 'r', encoding = 'utf-8-sig') as f:
    csv_reader = csv.reader(f)
    for row in csv_reader:
        row[0] = row[0][39:]
        row[0] = row[0][:-35]
        row[0] = row[0].replace('"','')
        image_link_lst.append(start_link + row[0])

print(image_link_lst)

with open(writefilename, 'w', encoding = 'utf-8-sig') as f:    
    write = csv.writer(f)
    write.writerows(image_link_lst)

import pandas as pd
df=pd.DataFrame(image_link_lst)
df.to_csv('image_links_final.csv', index=False)
```
